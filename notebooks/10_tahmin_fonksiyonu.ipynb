```python
import pandas as pd
import joblib
```

```python
model_path = "../models/balanced_logistic_regression.pkl"
```

```python
model = joblib.load(model_path)
```

```python
def predict_credit_risk(customer_data):
    customer_df = pd.DataFrame([customer_data])
    
    prediction = model.predict(customer_df)[0]
    prediction_proba = model.predict_proba(customer_df)[0]
    
    result = {
        "prediction": int(prediction),
        "prediction_label": "Kötü kredi riski" if prediction == 0 else "İyi kredi riski",
        "bad_credit_probability": float(round(prediction_proba[0], 3)),
        "good_credit_probability": float(round(prediction_proba[1], 3))
    }
    
    return result
```

```python
sample_customer = {
    "status": 1,
    "duration": 24,
    "credit_history": 2,
    "purpose": 3,
    "amount": 3000,
    "savings": 2,
    "employment_duration": 3,
    "installment_rate": 4,
    "personal_status_sex": 3,
    "other_debtors": 1,
    "present_residence": 2,
    "property": 2,
    "age": 35,
    "other_installment_plans": 3,
    "housing": 2,
    "number_credits": 1,
    "job": 3,
    "people_liable": 2,
    "telephone": 1,
    "foreign_worker": 2
}
```

```python
prediction_result = predict_credit_risk(sample_customer)
prediction_result
```

Bu bölümde tek müşteri için kredi riski tahmini yapan bir fonksiyon yazdım. Fonksiyon müşteri bilgilerini sözlük formatında alıyor, bu bilgiyi pandas DataFrame yapısına çeviriyor ve kaydedilmiş final model ile tahmin yapıyor.

Fonksiyon sonuç olarak tahmin sınıfını, okunabilir tahmin etiketini, kötü kredi olasılığını ve iyi kredi olasılığını döndürüyor. Böylece ileride Streamlit arayüzünde veya API tarafında aynı fonksiyon kolayca kullanılabilir.
