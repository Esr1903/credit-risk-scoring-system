```python
import pandas as pd
import joblib
```

```python
model_path = "../models/balanced_logistic_regression.pkl"
```

```python
model = joblib.load(model_path)
```

```python
sample_customer = pd.DataFrame([{
    "status": 1,
    "duration": 24,
    "credit_history": 2,
    "purpose": 3,
    "amount": 3000,
    "savings": 2,
    "employment_duration": 3,
    "installment_rate": 4,
    "personal_status_sex": 3,
    "other_debtors": 1,
    "present_residence": 2,
    "property": 2,
    "age": 35,
    "other_installment_plans": 3,
    "housing": 2,
    "number_credits": 1,
    "job": 3,
    "people_liable": 2,
    "telephone": 1,
    "foreign_worker": 2
}])
```

```python
sample_customer
```

```python
prediction = model.predict(sample_customer)
prediction
```

```python
prediction_label = "Kötü kredi riski" if prediction[0] == 0 else "İyi kredi riski"

prediction_label
```

```python
prediction_proba = model.predict_proba(sample_customer)
prediction_proba
```

Bunu şöyle okuyabiliriz:
0 sınıfı olasılığı = %68.3
1 sınıfı olasılığı = %31.7

Yani model bu müşteri için:
%68.3 kötü kredi riski
%31.7 iyi kredi riski
gibi bir değerlendirme yapmış.

```python
bad_credit_probability = prediction_proba[0][0]
good_credit_probability = prediction_proba[0][1]

print("Kötü kredi olasılığı:", round(bad_credit_probability, 3))
print("İyi kredi olasılığı:", round(good_credit_probability, 3))
```

Bu aşamada kaydettiğim final modeli kullanarak tek bir örnek müşteri için tahmin yaptım. Model bu müşteri için kötü kredi riski tahmini verdi. Tahmin olasılıklarına baktığımda kötü kredi olasılığı yaklaşık %68.3, iyi kredi olasılığı ise yaklaşık %31.7 çıktı.

Bu sonuç, kaydedilen modelin yeni bir müşteri verisi üzerinde kullanılabildiğini gösteriyor. Gerçek kullanımda kullanıcıdan alınan müşteri bilgileri bu formatta modele verilebilir ve model kredi riski tahmini üretebilir.
